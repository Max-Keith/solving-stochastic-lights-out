```python
import numpy as np
from matplotlib import pyplot as plt

from lightsout.stochastic.kernel import make_exp_kernel
from lightsout.stochastic.viz import show_action_heatmap

n = 5
actions = np.random.choice(n * n, size=5, replace=True)
# actions = [12]
P = make_exp_kernel(n=n, alpha=0.8, lam=0.0001)
show_action_heatmap(P, actions, n=n, cmap="magma")
plt.show()
```

```python
from lightsout.stochastic.viz import (
    show_difference_heatmap,
    show_entropy_heatmap,
)


show_difference_heatmap(P, actions, n)
show_entropy_heatmap(P, actions, n)
plt.show()
```
